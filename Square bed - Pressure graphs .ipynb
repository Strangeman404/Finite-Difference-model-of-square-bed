```python
import numpy, pandas
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
Data1 = numpy.asarray(pandas.read_excel('R_0.1.xlsx'))
Data2 = numpy.asarray(pandas.read_excel('R_0.2.xlsx'))
Data3 = numpy.asarray(pandas.read_excel('R_0.3.xlsx'))
Data4 = numpy.asarray(pandas.read_excel('R_0.4.xlsx'))
```

```python
Data1 = numpy.vstack(([0,298.15],Data1))
Data2 = numpy.vstack(([0,298.15],Data2))
Data3 = numpy.vstack(([0,298.15],Data3))
Data4 = numpy.vstack(([0,298.15],Data4))
```

```python
plt.figure()
plt.plot(Data1[:,0], Data1[:,1], label = '$R_i$ = 0.1 m', color = 'cyan')
plt.plot(Data2[:,0], Data2[:,1], label = '$R_i$ = 0.2 m', color = 'teal')
plt.plot(Data3[:,0], Data3[:,1], label = '$R_i$ = 0.3 m', color = 'steelblue')
plt.plot(Data4[:,0], Data4[:,1], label = '$R_i$ = 0.4 m', color = 'royalblue')

plt.legend(loc = 'lower right',prop={'size':15})
plt.xlabel('Length (m)',fontsize=15)
plt.ylabel('Temperature (K)',fontsize=15)
plt.axis((0,5,0,1200))

plt.xticks(fontsize=11, rotation=0)
plt.yticks(fontsize=11, rotation=0)

plt.savefig("Square bed gas.svg")
```
