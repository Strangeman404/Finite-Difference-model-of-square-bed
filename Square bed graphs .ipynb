```python
import numpy, pandas
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
Data1 = numpy.asarray(pandas.read_excel('Ri = 0.1, Ro = 0.2.xlsx'))
Data2 = numpy.asarray(pandas.read_excel('Ri = 0.1, Ro = 0.15.xlsx'))
Data3 = numpy.asarray(pandas.read_excel('Ri = 0.1, Ro = 0.25.xlsx'))
# Data4 = numpy.asarray(pandas.read_excel('Ri = 0.2, Ro = 0.3.xlsx'))
# Data5 = numpy.asarray(pandas.read_excel('Ri = 0.2, Ro = 0.25.xlsx'))
# Data6 = numpy.asarray(pandas.read_excel('Ri = 0.2, Ro = 0.35.xlsx'))
```

```python
plt.figure()
plt.plot(Data2[:,0], Data2[:,1], label = 'w = 0.05 m', color = 'cyan')
plt.plot(Data1[:,0], Data1[:,1], label = 'w = 0.10 m', color = 'teal')
plt.plot(Data3[:,0], Data3[:,1], label = 'w = 0.15 m', color = 'royalblue')
# plt.plot(Data4[:,0], Data4[:,1])
# plt.plot(Data5[:,0], Data5[:,1])
# plt.plot(Data6[:,0], Data6[:,1])

plt.legend(loc = 'lower right',prop={'size':15})
plt.xlabel('Time (s)',fontsize = 15)
plt.ylabel('Temperature (K)',fontsize = 15)
plt.axis((0,5000,0,1200))

plt.xticks(fontsize=11, rotation=0)
plt.yticks(fontsize=11, rotation=0)

plt.savefig("Square bed solids.svg")
```
